## 一、线性回归

### 1. **模型假设**
线性回归试图找到一个线性关系来预测目标变量 $ y $。假设模型为：

$$
y = w \cdot x + b
$$

其中：
- $ y $ 是目标变量（因变量）
- $ x $ 是输入特征（自变量）
- $ w $ 是回归系数（权重）
- $ b $ 是偏置项（截距）

### 2. **损失函数 (Mean Squared Error, MSE)**
为了最小化预测值与真实值之间的误差，线性回归通常使用均方误差（MSE）作为损失函数：

$$
J(w, b) = \frac{1}{2N} \sum_{i=1}^{N} (y_i - (w x_i + b))^2
$$

其中：
- $ N $ 是样本数
- $ y_i $ 是第 $ i $ 个样本的真实值
- $ \hat{y}_i = w x_i + b $ 是第 $ i $ 个样本的预测值

为什么是 $\frac{1}{2N}$ 呢？为了在求梯度时简化计算。

### 3. **梯度计算**
通过对损失函数 $ J(w, b) $ 分别对 $ w $ 和 $ b $ 求偏导数，得到梯度：

- 对 $ w $ 的偏导数（梯度）：

$$
\frac{\partial J}{\partial w} = -\frac{1}{N} \sum_{i=1}^{N} x_i (y_i - (w x_i + b))
$$

- 对 $ b $ 的偏导数（梯度）：

$$
\frac{\partial J}{\partial b} = -\frac{1}{N} \sum_{i=1}^{N} (y_i - (w x_i + b))
$$

### 4. **参数更新（梯度下降法）**
使用梯度下降法来优化参数 $ w $ 和 $ b $，每次更新规则为：

- 更新 $ w $ 和 $ b $：

$$
w = w - \alpha \cdot \frac{\partial J}{\partial w}
$$
$$
b = b - \alpha \cdot \frac{\partial J}{\partial b}
$$

其中：
- $ \alpha $ 是学习率，控制每次更新的步长。

### 5. **预测公式**
最终，通过训练得到的参数 $ w $ 和 $ b $，我们可以使用以下公式进行预测：

$$
\hat{y} = w \cdot x + b
$$

### 6. **模型评估 (MSE 和 R²)**
- **均方误差 (MSE)**：

$$
MSE = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2
$$

- **决定系数 (R²)**：

$$
R^2 = 1 - \frac{\sum_{i=1}^{N} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}
$$

其中 $ \bar{y} $ 是 $ y $ 的均值。


```python
import numpy as np 

class LinearRegression:
    def __init__(self, lr = 0.01, epochs = 1000):
        self.lr = lr
        self.epochs = epochs
        self.weights = None 
        self.bias = None 
    
    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weights = np.zeros(n_features)
        self.bias = 0 

        y = y.flatten()

        for _ in range(self.epochs):
            y_pred = np.dot(X, self.weights) + self.bias

            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
    
    def evaluate(self, X, y):
        y_pred = self.predict(X)
        mse = np.mean((y_pred-y) ** 2)
        r2 = 1 - (np.sum((y-y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2))
        return mse, r2
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100个数据点
y = 3 * X + 4 + np.random.randn(100, 1)  # 加入随机噪声

model = LinearRegression(lr=0.1, epochs=1000)
model.fit(X, y)

y_pred = model.predict(X)
mse, r2 = model.evaluate(X, y)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R^2: {r2}')
```

## 二、逻辑回归

### 1. **模型假设**
逻辑回归用于二分类问题，其目标是预测目标变量 $ y $ 属于某一类别的概率。假设模型为：

$$
P(y = 1 | x) = \sigma(w \cdot x + b)
$$

其中：
- $ P(y = 1 | x) $ 是给定输入 $ x $ 时，预测 $ y = 1 $ 的概率。
- $ \sigma(z) $ 是**sigmoid函数**，其定义为：

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

- $ w $ 是权重参数（系数）。
- $ b $ 是偏置项。

### 2. **损失函数 (Binary Cross-Entropy Loss)**
为了最小化预测值与真实值之间的误差，逻辑回归使用**二元交叉熵损失函数**（Binary Cross-Entropy Loss）作为损失函数：

$$
J(w, b) = - \frac{1}{N} \sum_{i=1}^{N} \left[y_i \log(P(y_i = 1 | x_i)) + (1 - y_i) \log(1 - P(y_i = 1 | x_i))\right]
$$

其中：
- $ N $ 是样本数
- $ y_i $ 是第 $ i $ 个样本的实际标签（0 或 1）
- $ P(y_i = 1 | x_i) = \sigma(w \cdot x_i + b) $ 是第 $ i $ 个样本的预测概率

### 3. **梯度计算**
通过对损失函数 $ J(w, b) $ 分别对 $ w $ 和 $ b $ 求偏导数，得到梯度：

- 对 $ w $ 的偏导数（梯度）：

$$
\frac{\partial J}{\partial w} = \frac{1}{N} \sum_{i=1}^{N} \left( \sigma(w \cdot x_i + b) - y_i \right) \cdot x_i
$$

- 对 $ b $ 的偏导数（梯度）：

$$
\frac{\partial J}{\partial b} = \frac{1}{N} \sum_{i=1}^{N} \left( \sigma(w \cdot x_i + b) - y_i \right)
$$

### 4. **参数更新（梯度下降法）**
使用梯度下降法来优化参数 $ w $ 和 $ b $，每次更新规则为：

- 更新 $ w $ 和 $ b $：

$$
w = w - \alpha \cdot \frac{\partial J}{\partial w}
$$
$$
b = b - \alpha \cdot \frac{\partial J}{\partial b}
$$

其中：
- $ \alpha $ 是学习率，控制每次更新的步长。

### 5. **预测公式**
最终，通过训练得到的参数 $ w $ 和 $ b $，我们可以使用以下公式进行预测：

$$
P(y = 1 | x) = \sigma(w \cdot x + b)
$$

根据阈值（通常为 0.5）来决定类别标签：
- 如果 $ P(y = 1 | x) \geq 0.5 $，预测 $ y = 1 $
- 否则，预测 $ y = 0 $

### 6. **模型评估**
- **准确率 (Accuracy)**：

$$
\text{Accuracy} = \frac{1}{N} \sum_{i=1}^{N} \mathbb{1} \left( \hat{y}_i = y_i \right)
$$

其中 $ \hat{y}_i $ 是预测值，$ \mathbb{1} $ 是指示函数。


```python
import numpy as np

class LogisticRegression:
    def __init__(self, lr=0.01, epochs=1000):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = None
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        y = y.flatten()

        for _ in range(self.epochs):
            linear_model = np.dot(X, self.weights) + self.bias
            y_pred = self.sigmoid(linear_model)

            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_pred = self.sigmoid(linear_model)
        return [1 if i >= 0.5 else 0 for i in y_pred]
    
    def evaluate(self, X, y):
        y_pred = self.predict(X)
        accuracy = np.mean(y_pred == y.flatten())
        return accuracy


# 数据
np.random.seed(42)
X = np.random.rand(100, 1)
y = (X > 0.5).astype(int)

# ✅ 修复：学习率改大！
model = LogisticRegression(lr=10, epochs=1000)
model.fit(X, y)

accuracy = model.evaluate(X, y)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

## 三、决策树

### 1. **模型假设**
决策树是一种用于分类和回归的监督学习算法。它通过对特征空间进行递归划分，生成一个树形结构，最终根据叶节点的标签进行预测。

每个内部节点表示一个特征的测试，每个分支表示该特征的一个可能值，叶节点表示最终的预测结果（类别标签或回归值）。

### 2. **信息增益与基尼指数**
决策树的训练过程是通过选择最佳的划分特征来不断地将数据集划分成更小的子集。为了选择最佳的划分标准，我们使用如下的指标：

#### **信息增益（Information Gain）**
信息增益衡量通过划分一个节点时，信息的不确定性减少了多少。信息增益的公式如下：

$$
IG(D, A) = H(D) - \sum_{v \in A} \frac{|D_v|}{|D|} H(D_v)
$$

其中：
- $ H(D) $ 是数据集 $ D $ 的熵。
- $ A $ 是特征集，$ v $ 是特征 $ A $ 的一个值。
- $ D_v $ 是根据特征 $ A $ 取值为 $ v $ 的数据子集。

#### **熵（Entropy）**
熵衡量了数据集的不确定性，公式如下：

$$
H(D) = - \sum_{i=1}^{k} p_i \log_2 p_i
$$

其中 $ p_i $ 是类别 $ i $ 在数据集 $ D $ 中的比例。

#### **基尼指数（Gini Index）**
基尼指数用于评估数据集的纯度，值越低表示纯度越高。计算公式为：

$$
Gini(D) = 1 - \sum_{i=1}^{k} p_i^2
$$

其中 $ p_i $ 是类别 $ i $ 在数据集 $ D $ 中的比例。

### 3. **选择最优划分**
在决策树的构建过程中，我们不断选择使得**信息增益最大**或**基尼指数最小**的特征进行划分。选择标准如下：

- **分类任务**：使用信息增益（ID3算法）或基尼指数（CART算法）。
- **回归任务**：使用均方误差（MSE）作为划分标准。

### 4. **树的构建过程**
1. **选择最佳特征**：通过计算信息增益或基尼指数，选择最佳特征进行划分。
2. **递归划分**：将数据集按选择的特征值分割成若干子集，递归地对每个子集继续选择特征进行划分，直到满足停止条件（例如最大深度或节点纯度达到阈值）。
3. **生成叶节点**：当满足停止条件时，节点会成为叶节点，叶节点会分配一个类别标签（分类任务）或回归值（回归任务）。

### 5. **决策树剪枝**
剪枝是通过去除不必要的节点来提高决策树的泛化能力。常见的剪枝方法有：
- **前剪枝**：在树的构建过程中，根据某些条件（例如最大深度、最小样本数等）停止划分。
- **后剪枝**：首先完全构建树，然后从叶节点开始逐步去除不重要的分支。


### ID3算法

```python
import numpy as np
from collections import Counter

# 决策树节点类
class Node:
    def __init__(
        self, feature=None, threshold=None, left=None, right=None, *, value=None
    ):
        # 内部节点
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        # 叶节点（最终类别）
        self.value = value

    def is_leaf(self):
        return self.value is not None

# 决策树分类器
class DecisionTree:
    def __init__(self, min_samples_split=2, max_depth=100):
        self.min_samples_split = min_samples_split  # 最小划分样本数
        self.max_depth = max_depth                  # 最大深度
        self.root = None                            # 根节点

    # ******************** 核心公式实现 ********************
    # 1. 计算熵
    def _entropy(self, y):
        hist = np.bincount(y)
        ps = hist / len(y)
        return -np.sum([p * np.log2(p) for p in ps if p > 0])

    # 2. 计算信息增益
    def _information_gain(self, y, X_column, threshold):
        # 父节点熵
        parent_entropy = self._entropy(y)

        # 划分左右子集
        left_idx = X_column <= threshold
        right_idx = X_column > threshold
        n = len(y)
        n_l, n_r = np.sum(left_idx), np.sum(right_idx)
        if n_l == 0 or n_r == 0:
            return 0

        # 子节点熵
        e_l = self._entropy(y[left_idx])
        e_r = self._entropy(y[right_idx])

        # 信息增益
        child_entropy = (n_l / n) * e_l + (n_r / n) * e_r
        ig = parent_entropy - child_entropy
        return ig

    # ******************** 树构建逻辑 ********************
    # 选择最优划分特征 + 阈值
    def _best_split(self, X, y, features):
        best_gain = -1
        split_idx, split_thresh = None, None

        for idx in features:
            X_col = X[:, idx]
            thresholds = np.unique(X_col)  # 所有可能划分点

            for thresh in thresholds:
                gain = self._information_gain(y, X_col, thresh)

                if gain > best_gain:
                    best_gain = gain
                    split_idx = idx
                    split_thresh = thresh

        return split_idx, split_thresh

    # 递归生成树
    def _grow_tree(self, X, y, depth=0):
        n_samples, n_feats = X.shape
        n_labels = len(np.unique(y))

        # 停止条件
        if (depth >= self.max_depth 
            or n_labels == 1 
            or n_samples < self.min_samples_split):
            leaf_value = self._most_common_label(y)
            return Node(value=leaf_value)

        # 随机选择特征（防止过拟合）
        feat_idxs = np.random.choice(n_feats, n_feats, replace=False)
        best_feat, best_thresh = self._best_split(X, y, feat_idxs)

        # 划分数据
        left_idxs = X[:, best_feat] <= best_thresh
        right_idxs = X[:, best_feat] > best_thresh

        # 递归
        left = self._grow_tree(X[left_idxs], y[left_idxs], depth + 1)
        right = self._grow_tree(X[right_idxs], y[right_idxs], depth + 1)
        return Node(best_feat, best_thresh, left, right)

    # 叶节点：最多类别
    def _most_common_label(self, y):
        counter = Counter(y)
        return counter.most_common(1)[0][0]

    # ******************** 训练/预测/评估 ********************
    def fit(self, X, y):
        self.root = self._grow_tree(X, y)

    def _traverse_tree(self, x, node):
        if node.is_leaf():
            return node.value
        if x[node.feature] <= node.threshold:
            return self._traverse_tree(x, node.left)
        return self._traverse_tree(x, node.right)

    def predict(self, X):
        return np.array([self._traverse_tree(x, self.root) for x in X])

    def evaluate(self, X, y):
        y_pred = self.predict(X)
        acc = np.sum(y_pred == y) / len(y)
        return acc
```

```python
# 测试数据（和你逻辑回归完全一样）
np.random.seed(42)
X = np.random.rand(100, 1)
y = (X > 0.5).astype(int).flatten()

# 训练决策树
dt = DecisionTree(max_depth=10, min_samples_split=2)
dt.fit(X, y)

# 评估
acc = dt.evaluate(X, y)
print(f"决策树准确率: {acc * 100:.2f}%")
```

### CART算法

```python
import numpy as np
from collections import Counter

# ---------------------
# 1. 树节点（和之前一样）
# ---------------------
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature      # 用哪个特征划分
        self.threshold = threshold  # 划分阈值
        self.left = left            # 左子树
        self.right = right          # 右子树
        self.value = value          # 叶子节点的值

    def is_leaf(self):
        return self.value is not None

# ---------------------
# 2. CART 决策树（基尼系数）
# ---------------------
class CARTClassifier:
    def __init__(self, max_depth=10, min_samples_split=2):
        self.max_depth = max_depth              # 树最大深度
        self.min_samples_split = min_samples_split  # 最少样本才允许分
        self.root = None

    # ---------------------
    # 🔑 CART 核心：基尼系数
    # ---------------------
    def _gini(self, y):
        counts = np.bincount(y)
        p = counts / len(y)
        return 1 - np.sum(p ** 2)

    # ---------------------
    # 计算划分后的基尼增益
    # ---------------------
    def _gini_gain(self, y, left_mask, right_mask):
        n = len(y)
        n_left = np.sum(left_mask)
        n_right = np.sum(right_mask)

        if n_left == 0 or n_right == 0:
            return 0

        gini_left = self._gini(y[left_mask])
        gini_right = self._gini(y[right_mask])

        child_gini = (n_left / n) * gini_left + (n_right / n) * gini_right
        return self._gini(y) - child_gini

    # ---------------------
    # 遍历所有特征+阈值，找最优划分（CART 核心）
    # ---------------------
    def _best_split(self, X, y):
        best_gain = -1
        best_idx = None
        best_thresh = None

        n_samples, n_features = X.shape

        # 遍历每个特征
        for idx in range(n_features):
            X_col = X[:, idx]
            thresholds = np.unique(X_col)  # 所有可能的划分点

            # 遍历每个阈值
            for thresh in thresholds:
                left_mask = X_col <= thresh
                right_mask = X_col > thresh

                # 计算基尼增益
                gain = self._gini_gain(y, left_mask, right_mask)

                if gain > best_gain:
                    best_gain = gain
                    best_idx = idx
                    best_thresh = thresh

        return best_idx, best_thresh

    # ---------------------
    # 叶子节点：最多类别
    # ---------------------
    def _most_common(self, y):
        return Counter(y).most_common(1)[0][0]

    # ---------------------
    # 递归构建 CART 二叉树
    # ---------------------
    def _build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))

        # 停止条件
        if (depth >= self.max_depth 
            or n_samples < self.min_samples_split 
            or n_classes == 1):
            leaf_value = self._most_common(y)
            return Node(value=leaf_value)

        # 找最优划分
        best_idx, best_thresh = self._best_split(X, y)

        # 划分数据
        left_mask = X[:, best_idx] <= best_thresh
        right_mask = X[:, best_idx] > best_thresh

        # 递归建左右子树
        left = self._build_tree(X[left_mask], y[left_mask], depth + 1)
        right = self._build_tree(X[right_mask], y[right_mask], depth + 1)

        return Node(best_idx, best_thresh, left, right)

    # ---------------------
    # 训练
    # ---------------------
    def fit(self, X, y):
        self.root = self._build_tree(X, y)

    # ---------------------
    # 单样本预测
    # ---------------------
    def _traverse(self, x, node):
        if node.is_leaf():
            return node.value
        if x[node.feature] <= node.threshold:
            return self._traverse(x, node.left)
        else:
            return self._traverse(x, node.right)

    # ---------------------
    # 批量预测
    # ---------------------
    def predict(self, X):
        return np.array([self._traverse(x, self.root) for x in X])

    # ---------------------
    # 评估准确率
    # ---------------------
    def score(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == y)
```

```python
# 测试数据
np.random.seed(42)
X = np.random.rand(100, 1)
y = (X > 0.5).astype(int).flatten()

# CART 模型
model = CARTClassifier(max_depth=3)
model.fit(X, y)

acc = model.score(X, y)
print(f"CART 决策树准确率: {acc*100:.2f}%")
```

## 四、支持向量机(SVM)

### 1. **模型假设**
支持向量机（SVM）是一种用于分类和回归的监督学习算法，特别适用于高维数据。其核心思想是通过在特征空间中找到一个最大化间隔的超平面来进行分类。

对于一个线性可分的二分类问题，SVM的目标是找到一个超平面：

$$
w \cdot x + b = 0
$$

其中：
- $ w $ 是超平面的法向量（权重向量）。
- $ b $ 是偏置项。
- $ x $ 是输入数据。

### 2. **最大化间隔**
为了最大化分类的间隔，SVM 的目标是找到一个最佳的超平面，该平面能够使得两类样本之间的间隔最大化。对于支持向量机，我们的目标是最大化如下的间隔：

$$
\text{margin} = \frac{2}{\|w\|}
$$

因此，SVM 的目标就是最小化目标函数：

$$
\min_{w, b} \frac{1}{2} \|w\|^2
$$

同时，要求每个样本点 $ (x_i, y_i) $ 满足以下约束条件：

$$
y_i (w \cdot x_i + b) \geq 1, \quad i = 1, 2, \dots, N
$$

其中：
- $ y_i \in \{-1, +1\} $ 是样本点的标签。
- $ x_i $ 是第 $ i $ 个样本点。

### 3. **目标函数与约束条件**
SVM 的优化问题可以写成如下的形式：

$$
\min_{w, b} \frac{1}{2} \|w\|^2
$$
$$
\text{subject to } y_i (w \cdot x_i + b) \geq 1, \quad i = 1, 2, \dots, N
$$

### 4. **拉格朗日对偶问题**
为了求解这个优化问题，我们引入拉格朗日乘子法。将约束条件引入拉格朗日乘子后，原问题转化为对偶问题：

$$
\mathcal{L}(w, b, \alpha) = \frac{1}{2} \|w\|^2 - \sum_{i=1}^{N} \alpha_i \left( y_i (w \cdot x_i + b) - 1 \right)
$$

其中，$ \alpha_{i} \geq 0 $ 是拉格朗日乘子。

通过求解对偶问题，可以得到最优的 $ w $ 和 $ b $。

对于凸优化问题（SVM 是凸的），最优解必须满足 Karush–Kuhn–Tucker (KKT) 条件，其中最关键的一条就是：
互补松弛条件（Complementary Slackness）：
$$
\alpha_{i} \cdot g_{i}(w^{*}, b^{*}) = 0, \forall i
$$
也就是：
$$
\alpha_{i} \cdot [1-y_{i}(w^{*T}x_{i}+b^{*})] = 0, \forall i
$$
这意味着：对于每一个样本$i$, 要么$\alpha_{i} = 0$，要么约束取等号（即正好在间隔边界上），此时这类在边界上的样本点叫做支持向量。


### 5. **支持向量机的核技巧**
对于非线性可分的情况，SVM 可以通过使用**核函数**将数据映射到更高维的特征空间，使得数据在高维空间中变得线性可分。常用的核函数包括：
- **线性核函数**：$ K(x, x') = x \cdot x' $
- **多项式核函数**：$ K(x, x') = (x \cdot x' + c)^d $
- **径向基核函数（RBF）**：$ K(x, x') = \exp(-\gamma \|x - x'\|^2) $

在使用核函数时，优化问题变为：

$$
\min_{\alpha} \sum_{i=1}^{N} \alpha_i - \frac{1}{2} \sum_{i,j=1}^{N} \alpha_i \alpha_j y_i y_j K(x_i, x_j)
$$

### 6. **支持向量**
在最终的决策函数中，只有那些满足支持向量条件的样本点对最终的决策面有贡献。支持向量是指那些位于分类边界附近的点，且对应的拉格朗日乘子 $ \alpha_i > 0 $。

最终的分类函数为：

$$
f(x) = \text{sign}\left( \sum_{i=1}^{N} \alpha_i y_i K(x_i, x) + b \right)
$$

其中 $ \alpha_i $ 是通过对偶问题得到的拉格朗日乘子，$ K(x_i, x) $ 是核函数，$ b $ 是偏置。


```python
import numpy as np

# 线性核函数
def linear_kernel(x1, x2):
    return np.dot(x1, x2)

# SVM 类（SMO算法实现）
class SVM:
    def __init__(self, C=1.0, kernel=linear_kernel, tol=1e-3, max_iter=1000):
        self.C = C          # 惩罚参数
        self.kernel = kernel
        self.tol = tol      # 收敛阈值
        self.max_iter = max_iter
        self.alpha = None   # 拉格朗日乘子
        self.b = 0.0        # 偏置项
        self.X = None
        self.y = None

    def fit(self, X, y):
        self.X = X
        self.y = y
        m, n = X.shape
        self.alpha = np.zeros(m)  # 初始化alpha
        iter_num = 0

        # SMO 迭代优化
        while iter_num < self.max_iter:
            alpha_prev = np.copy(self.alpha)
            for i in range(m):
                # 随机选另一个变量 j
                j = np.random.choice([k for k in range(m) if k != i])

                # 计算误差
                error_i = self._predict_single(X[i]) - y[i]
                error_j = self._predict_single(X[j]) - y[j]

                # 计算 eta
                eta = self.kernel(X[i], X[i]) + self.kernel(X[j], X[j]) - 2 * self.kernel(X[i], X[j])
                if eta <= 0:
                    continue

                # 保存旧alpha
                alpha_i_old = self.alpha[i]
                alpha_j_old = self.alpha[j]

                # 计算上下界 L, H
                if y[i] != y[j]:
                    L = max(0, alpha_j_old - alpha_i_old)
                    H = min(self.C, self.C + alpha_j_old - alpha_i_old)
                else:
                    L = max(0, alpha_i_old + alpha_j_old - self.C)
                    H = min(self.C, alpha_i_old + alpha_j_old)

                if L == H:
                    continue

                # 更新 alpha_j
                self.alpha[j] = alpha_j_old + y[j] * (error_i - error_j) / eta
                # 裁剪 alpha_j
                self.alpha[j] = np.clip(self.alpha[j], L, H)

                # 更新 alpha_i
                self.alpha[i] = alpha_i_old + y[i] * y[j] * (alpha_j_old - self.alpha[j])

                # 更新阈值 b
                b1 = self.b - error_i - y[i] * (self.alpha[i] - alpha_i_old) * self.kernel(X[i], X[i]) - y[j] * (self.alpha[j] - alpha_j_old) * self.kernel(X[i], X[j])
                b2 = self.b - error_j - y[i] * (self.alpha[i] - alpha_i_old) * self.kernel(X[i], X[j]) - y[j] * (self.alpha[j] - alpha_j_old) * self.kernel(X[j], X[j])
                
                if 0 < self.alpha[i] < self.C:
                    self.b = b1
                elif 0 < self.alpha[j] < self.C:
                    self.b = b2
                else:
                    self.b = (b1 + b2) / 2

            # 判断收敛
            if np.linalg.norm(self.alpha - alpha_prev) < self.tol:
                break
            iter_num += 1

    # 预测单个样本
    def _predict_single(self, x):
        return np.sum(self.alpha * self.y * self.kernel(self.X, x)) + self.b

    # 预测批量样本
    def predict(self, X):
        return np.sign(np.array([self._predict_single(x) for x in X]))

# 测试
if __name__ == "__main__":
    X = np.array([[2.7, 1.7], [1.5, 2.5], [3.5, 4.2], [4.1, 3.1], [2.1, 1.2]])
    y = np.array([1, 1, -1, -1, 1])

    model = SVM(C=1.0)
    model.fit(X, y)
    pred = model.predict(X)
    print("真实标签:", y)
    print("预测结果:", pred.astype(int))
```

## 五、KNN

**KNN（K 近邻算法）** 是最简单、最直观的**懒惰学习**（lazy learning）算法。  
它**没有显式的训练过程**，只是把所有训练样本“记住”，预测时才去计算距离。

---

### 1. KNN 核心思想

**“物以类聚”**：一个样本的类别，由它**最近的 K 个邻居**投票决定（分类）或取平均（回归）。

- **无参数模型**：不需要学习 w、b、α 等参数
- **非线性**：天然支持非线性决策边界
- **缺点**：预测慢（每次都要遍历所有样本）、对噪声敏感、维度灾难

---

### 2. 核心公式（必须掌握）

#### 2.1 距离度量（最常用 4 种）

##### （1）欧氏距离（Euclidean Distance，默认推荐）
$$
d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

##### （2）曼哈顿距离（Manhattan Distance）
$$
d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} |x_i - y_i|
$$

##### （3）闵可夫斯基距离（Minkowski Distance，通用形式）
$$
d(\mathbf{x}, \mathbf{y}) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{1/p}
$$
- p=2 → 欧氏距离  
- p=1 → 曼哈顿距离  
- p=∞ → 切比雪夫距离

##### （4）余弦相似度（常用于文本、高维稀疏数据）
$$
\cos(\theta) = \frac{\mathbf{x} \cdot \mathbf{y}}{||\mathbf{x}|| \cdot ||\mathbf{y}||}
$$
（KNN 通常用距离，所以取 1 - cos(θ) 作为距离）

#### 2.2 预测公式

##### 分类任务（Classification）
$$
\hat{y} = \text{mode} \Big\{ y_i \mid i \in \text{TopK nearest neighbors} \Big\}
$$
即 **K 个邻居中出现次数最多的类别**（多数投票）。

##### 回归任务（Regression）
$$
\hat{y} = \frac{1}{K} \sum_{i \in \text{TopK}} y_i
$$
即 **K 个邻居的目标值平均值**。

#### 2.3 加权版本（更常用，效果更好）
给距离近的邻居更大权重：

分类：
$$
\hat{y} = \arg\max_c \sum_{i \in \text{TopK}} w_i \cdot \mathbb{I}(y_i = c)
$$

回归：
$$
\hat{y} = \frac{\sum_{i \in \text{TopK}} w_i \cdot y_i}{\sum_{i \in \text{TopK}} w_i}
$$

其中权重常用：
$$
w_i = \frac{1}{d(\mathbf{x}, \mathbf{x}_i) + \epsilon} \quad (\epsilon \text{防止除0})
$$

### 分类任务

```python
import numpy as np
from collections import Counter

class SimpleKNN:
    def __init__(self, k=5, weights='uniform', metric='euclidean'):
        self.k = k
        self.weights = weights
        self.metric = metric
        self.X_train = None
        self.y_train = None

    def _distance(self, x1, x2):
        if self.metric == 'euclidean':
            return np.sqrt(np.sum((x1 - x2)**2, axis=1))
        elif self.metric == 'manhattan':
            return np.sum(np.abs(x1 - x2), axis=1)
        elif self.metric == 'cosine':
            return 1 - np.dot(x1, x2.T) / (np.linalg.norm(x1, axis=1) * np.linalg.norm(x2, axis=1) + 1e-8)

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def predict(self, X):
        X = np.array(X)
        predictions = []
        for x in X:
            # 计算所有训练样本的距离
            distances = self._distance(self.X_train, x.reshape(1, -1))
            
            # 找到最近的 k 个索引
            k_idx = np.argsort(distances)[:self.k]
            k_labels = self.y_train[k_idx]
            k_dist = distances[k_idx]
            
            if self.weights == 'uniform':
                # 多数投票
                pred = Counter(k_labels).most_common(1)[0][0]
            else:  # distance 加权
                weights = 1 / (k_dist + 1e-8)
                weighted_votes = {}
                for label, w in zip(k_labels, weights):
                    weighted_votes[label] = weighted_votes.get(label, 0) + w
                pred = max(weighted_votes, key=weighted_votes.get)
            
            predictions.append(pred)
        return np.array(predictions)

# 测试
if __name__ == "__main__":
    from sklearn.datasets import load_iris
    iris = load_iris()
    X, y = iris.data[:100], iris.target[:100]  # 二分类简化测试
    
    knn = SimpleKNN(k=5, weights='distance')
    knn.fit(X, y)
    print("预测前5个:", knn.predict(X[:5]))
    print("真实前5个:", y[:5])
```

## 六、随机森林（Random Forest）

**随机森林（Random Forest）** 是 **Bagging（Bootstrap Aggregating）** + **随机特征选择** 的集成学习算法。
它是目前工业界最常用、最稳健的经典算法之一，被称为**不需要调参也能跑得很好**的算法。

---

### 1. 随机森林核心思想（一句话记住）
**一群决策树，每棵树都用不同数据 + 不同特征训练，最后集体投票 / 平均**。
- **强泛化能力**：单棵决策树容易过拟合，森林通过“民主投票”大幅降低方差
- **抗噪声、抗过拟合**：天然自带 OOB（Out-of-Bag）验证
- **并行训练**：每棵树独立训练，可完全并行

---

### 2. 核心公式（必须掌握）
#### 2.1 分类任务预测公式（多数投票）
$$
\hat{y} = \text{mode} \Big\{ T_1(\mathbf{x}), T_2(\mathbf{x}), \dots, T_B(\mathbf{x}) \Big\}
$$
其中：
- $B$ = 森林中树的数量（n_estimators）
- $T_b(\mathbf{x})$ = 第 $b$ 棵决策树对样本 $\mathbf{x}$ 的预测
- $\text{mode}$ = 出现次数最多的类别

#### 2.2 回归任务预测公式（平均）
$$
\hat{y} = \frac{1}{B} \sum_{b=1}^{B} T_b(\mathbf{x})
$$

#### 2.3 每棵树的训练过程（Bootstrap + 随机特征）
1. **Bootstrap 采样**（有放回抽样）
   从 $N$ 个样本中随机抽取 $N$ 个（允许重复），平均约 **37%** 样本不会被抽到（称为 OOB 样本）。
2. **随机特征选择**（节点分裂时）
   每个节点分裂时，只从全部 $M$ 个特征中**随机挑选** $m$ 个：
   - 分类：$m = \sqrt{M}$
   - 回归：$m = M/3$

   目的：让每棵决策树长得不一样，避免使用相同的特征，从而使得集成后泛化能力加强。
3. **不剪枝**：每棵树都长到最大深度（或节点纯净）

#### 2.4 特征重要性
基于 **Gini 系数下降总和**（分类）/ MSE 下降（回归）：
$$
\text{Importance}(j) = \sum_{b=1}^{B} \sum_{t \in T_b} \Delta \text{Gini}_{t,j} \cdot \frac{N_t}{N}
$$
`sklearn` 中用 `feature_importances_` 直接输出（归一化后和为 1）。

---

### 4. 关键超参数说明
| 参数 | 作用 | 推荐值 |
|------|------|--------|
| `n_estimators` | 森林中树的数量 | 100~200（越大越稳） |
| `max_depth` | 树的最大深度 | None（不剪枝） |
| `max_features` | 每次分裂随机选的特征数 | 分类`sqrt`，回归`1/3` |
| `bootstrap` | 是否使用 Bootstrap 采样 | True（开启随机森林） |
| `oob_score` | 是否使用 OOB 验证 | True |
| `n_jobs` | 并行训练 | -1（使用所有CPU） |

---

### 5. 随机森林为什么效果好？
1. **样本随机**：Bootstrap 采样，降低数据相关性
2. **特征随机**：节点分裂随机选特征，进一步降低相关性
3. **集成投票**：多棵树独立预测，集体决策，大幅降低过拟合
4. **自带验证**：OOB 得分无需划分验证集，直接评估模型

---

#### 总结
1. 随机森林 = **Bagging + 随机特征 + 决策树**
2. 分类用**投票**，回归用**平均**
3. 自带 **OOB 验证** 和 **特征重要性**
4. 工业界最实用、最稳健的 baseline 模型

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier  # 用 sklearn 单棵树简化实现
from sklearn.utils import resample
from collections import Counter

class SimpleRandomForest:
    def __init__(self, n_trees=10, max_features='sqrt', max_depth=None):
        self.n_trees = n_trees
        self.max_features = max_features
        self.max_depth = max_depth
        self.trees = []
        self.feature_subsets = []

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        n_features = X.shape[1]
        
        if self.max_features == 'sqrt':
            m = int(np.sqrt(n_features))
        elif self.max_features == 'log2':
            m = int(np.log2(n_features))
        else:
            m = n_features

        for i in range(self.n_trees):
            # 1. Bootstrap 采样
            X_boot, y_boot = resample(X, y, n_samples=len(X), random_state=i)
            
            # 2. 随机选特征子集
            feat_idx = np.random.choice(n_features, m, replace=False)
            self.feature_subsets.append(feat_idx)
            
            # 3. 训练单棵决策树（不剪枝）
            tree = DecisionTreeClassifier(
                max_depth=self.max_depth,
                max_features=None,   # 已提前选好特征
                random_state=i
            )
            tree.fit(X_boot[:, feat_idx], y_boot)
            
            self.trees.append(tree)
    
    def predict(self, X):
        X = np.array(X)
        predictions = np.zeros((X.shape[0], self.n_trees), dtype=int)
        
        for i, tree in enumerate(self.trees):
            feat_idx = self.feature_subsets[i]
            pred = tree.predict(X[:, feat_idx])
            predictions[:, i] = pred
        
        # 多数投票
        final_pred = []
        for row in predictions:
            final_pred.append(Counter(row).most_common(1)[0][0])
        return np.array(final_pred)

# 测试
if __name__ == "__main__":
    from sklearn.datasets import load_iris
    iris = load_iris()
    X, y = iris.data[:100], iris.target[:100]  # 二分类简化
    
    rf = SimpleRandomForest(n_trees=20, max_features='sqrt')
    rf.fit(X, y)
    print("预测前5个:", rf.predict(X[:5]))
    print("真实前5个:", y[:5])
```

## 七、XGBoost（eXtreme Gradient Boosting）

### 1. XGBoost 核心定位与基础思想
XGBoost 是**极端梯度提升树**，属于**Boosting**集成算法，是梯度提升决策树（GBDT）的工程与理论极致优化版本。
与随机森林（Bagging）并行训练多棵树不同，XGBoost 采用**串行训练**：**一棵树一棵树依次构建，每棵新树都学习之前所有树的预测残差/梯度，最终将所有树的输出累加得到预测结果**。

核心特性：
- 加法模型：最终预测为所有决策树输出的加权和
- 二阶泰勒展开：使用一阶梯度+二阶梯度优化，比传统GBDT更精准、收敛更快
- 内置正则化：从结构上约束模型，从根源防止过拟合
- 高效分裂算法与工程优化：支持缺失值处理、特征并行、缓存优化

---
### 2. 核心数学公式（逐行推导+详细解释）
#### 2.1 整体加法模型定义
XGBoost 由 $K$ 棵独立的CART决策树组成，最终预测值为所有树输出之和：
$$
\hat{y}_i = \sum_{k=1}^K f_k(x_i), \quad f_k \in \mathcal{F}
$$

**参数解释**：
- $\hat{y}_i$：第 $i$ 个样本的模型预测值
- $x_i$：第 $i$ 个样本的特征向量
- $K$：集成模型中决策树的总数量
- $f_k$：第 $k$ 棵决策树（映射样本到叶子节点权重）
- $\mathcal{F}$：所有可能的CART决策树构成的函数空间

---
#### 2.2 第 $t$ 轮迭代目标函数（核心）
XGBoost 采用**前向分步训练**：训练到第 $t$ 棵树时，保留前 $t-1$ 棵树不变，仅优化第 $t$ 棵树。
目标函数 = 训练损失 + 正则化项（结构惩罚）：
$$
Obj^{(t)} = \sum_{i=1}^n l(y_i, \hat{y}_i^{(t)}) + \sum_{k=1}^t \Omega(f_k)
$$

**分项解释**：
1. 损失项 $\sum_{i=1}^n l(y_i, \hat{y}_i^{(t)})$
   - 衡量模型预测值与真实标签的误差
   - 分类任务常用：对数损失、交叉熵损失
   - 回归任务常用：平方损失
2. 正则项 $\sum_{k=1}^t \Omega(f_k)$
   - 惩罚复杂的树结构，防止过拟合
   - 是XGBoost优于传统GBDT的关键设计

---
#### 2.3 树结构正则项公式
XGBoost 从**叶子节点数量**和**叶子节点权重**两方面约束树结构：
$$
\Omega(f) = \gamma T + \frac{1}{2} \lambda \sum_{j=1}^T w_j^2
$$

**参数解释**：
- $\Omega(f)$：单棵决策树的结构复杂度惩罚
- $T$：当前树的**叶子节点总数**
- $w_j$：第 $j$ 个叶子节点的输出权重
- $\gamma$：叶子节点数惩罚系数（值越大，树越精简）
- $\lambda$：权重L2正则系数（值越大，权重越平滑）

---
#### 2.4 二阶泰勒展开（XGBoost 灵魂）
为了高效优化第 $t$ 棵树，将损失函数在 $\hat{y}_i^{(t-1)}$ 处做**二阶泰勒展开**（传统GBDT仅用一阶）：
$$
Obj^{(t)} \approx \sum_{i=1}^n \left[ l(y_i, \hat{y}_i^{(t-1)}) + g_i f_t(x_i) + \frac{1}{2} h_i f_t(x_i)^2 \right] + \Omega(f_t)
$$

**梯度定义**：
- 一阶梯度 $g_i = \frac{\partial l(y_i, \hat{y}_i^{(t-1)})}{\partial \hat{y}_i^{(t-1)}}$
- 二阶梯度 $h_i = \frac{\partial^2 l(y_i, \hat{y}_i^{(t-1)})}{\partial \hat{y}_i^{(t-1)}}$

**简化目标函数**：
$l(y_i, \hat{y}_i^{(t-1)})$ 是常数项，优化时可直接去除，最终第 $t$ 棵树的优化目标：
$$
Obj^{(t)} = \sum_{i=1}^n \left[ g_i f_t(x_i) + \frac{1}{2} h_i f_t(x_i)^2 \right] + \Omega(f_t)
$$

---
#### 2.5 叶子节点最优权重公式
当树的结构固定时，可直接求出**最优叶子权重**，使目标函数最小：
$$
w_j^* = -\frac{\sum_{i \in I_j} g_i}{\sum_{i \in I_j} h_i + \lambda}
$$

**参数解释**：
- $w_j^*$：第 $j$ 个叶子节点的最优输出权重
- $I_j$：落在第 $j$ 个叶子节点上的所有样本集合
- 分子：叶子节点内样本一阶梯度之和
- 分母：二阶梯度之和 + 正则项（保证分母不为0）

---
#### 2.6 节点分裂增益公式（分裂依据）
决策树分裂时，计算分裂带来的目标函数下降值（增益），增益越大分裂效果越好：
$$
\text{Gain} = \frac{1}{2} \left[ \frac{G_L^2}{H_L+\lambda} + \frac{G_R^2}{H_R+\lambda} - \frac{G^2}{H+\lambda} \right] - \gamma
$$

**参数解释**：
- $G_L, H_L$：左子节点梯度和、二阶梯度和
- $G_R, H_R$：右子节点梯度和、二阶梯度和
- $G, H$：分裂前父节点总梯度和、总二阶梯度和
- $\gamma$：分裂惩罚项（增益大于$\gamma$才允许分裂）

**核心规则**：
$\text{Gain} > 0$ → 执行节点分裂
$\text{Gain} \le 0$ → 不分裂，停止生长

---
### 3. 核心原理总结
1. **训练逻辑**：串行构建决策树，每棵树拟合残差/梯度
2. **优化核心**：二阶泰勒展开 + 结构正则化，兼顾精度与泛化能力
3. **分裂规则**：基于增益公式选择最优特征与分裂点
4. **预测逻辑**：所有树的叶子权重累加，得到最终预测结果

#### XGBoost 最优叶子权重 + 分裂增益（2.5、2.6）完整推导过程

#### 一、前提：第 t 棵树的简化目标函数
二阶泰勒展开 + 正则化后，目标函数为：
$$
Obj = \sum_{i=1}^n \left[ g_i f_t(x_i) + \frac{1}{2} h_i f_t(x_i)^2 \right] + \gamma T + \frac{1}{2} \lambda \sum_{j=1}^T w_j^2
$$

其中：
- $f_t(x_i) = w_j$：样本 $i$ 落在叶子 $j$，输出为 $w_j$
- $G_j = \sum_{i \in I_j} g_i$：叶子 $j$ 内一阶梯度总和
- $H_j = \sum_{i \in I_j} h_i$：叶子 $j$ 内二阶梯度总和
- $T$：叶子数量
- $\gamma, \lambda$：正则系数

---

#### 二、按叶子重新整理目标函数
同一个叶子内样本共享同一个 $w_j$，因此可按叶子求和：
$$
Obj = \sum_{j=1}^T \left[ G_j w_j + \frac{1}{2} H_j w_j^2 \right] + \gamma T + \frac{1}{2} \lambda \sum_{j=1}^T w_j^2
$$

合并 $w_j^2$ 项：
$$
Obj = \sum_{j=1}^T \left[ G_j w_j + \frac{1}{2} (H_j + \lambda) w_j^2 \right] + \gamma T
$$

---

#### 三、推导最优叶子权重 $w_j^*$
目标函数对 $w_j$ 求导并令导数为 0：

$$
\frac{\partial Obj}{\partial w_j} = G_j + (H_j + \lambda) w_j = 0
$$

解得：

$$
w_j^* = -\frac{G_j}{H_j + \lambda}
$$

---

#### 四、将最优权重代回目标函数
将 $w_j^*$ 代入 $Obj$：

$$
G_j w_j^* = -\frac{G_j^2}{H_j + \lambda}
$$

$$
\frac{1}{2}(H_j+\lambda) (w_j^*)^2 = \frac{1}{2} \cdot \frac{G_j^2}{H_j + \lambda}
$$

相加得单叶子最小目标值：

$$
-\frac{G_j^2}{H_j + \lambda} + \frac{1}{2} \frac{G_j^2}{H_j + \lambda} = -\frac{1}{2} \frac{G_j^2}{H_j + \lambda}
$$

整体最小目标函数：

$$
Obj_{min} = -\frac{1}{2} \sum_{j=1}^T \frac{G_j^2}{H_j + \lambda} + \gamma T
$$

---

#### 五、推导分裂增益 Gain
##### 分裂前（1 个叶子）
$$
Obj_{before} = -\frac{1}{2} \cdot \frac{G^2}{H + \lambda} + \gamma
$$

##### 分裂后（左、右两个叶子）
$$
Obj_{after} = -\frac{1}{2} \left( \frac{G_L^2}{H_L+\lambda} + \frac{G_R^2}{H_R+\lambda} \right) + 2\gamma
$$

##### 增益 = 分裂前 - 分裂后（损失下降量）
$$
Gain = Obj_{before} - Obj_{after}
$$

展开：
$$
Gain = \frac{1}{2} \left[ \frac{G_L^2}{H_L+\lambda} + \frac{G_R^2}{H_R+\lambda} - \frac{G^2}{H+\lambda} \right] - \gamma
$$

---

#### 最终结论
1. **最优叶子权重**
$$
w_j^* = -\frac{G_j}{H_j + \lambda}
$$

2. **分裂增益**
$$
Gain = \frac{1}{2} \left[ \frac{G_L^2}{H_L+\lambda} + \frac{G_R^2}{H_R+\lambda} - \frac{G^2}{H+\lambda} \right] - \gamma
$$

```python
import numpy as np

# 单棵 XGBoost 树（CART）
class XGBTree:
    def __init__(self, max_depth=3, reg_lambda=1, gamma=0):
        self.max_depth = max_depth
        self.reg_lambda = reg_lambda  # L2 正则
        self.gamma = gamma            # 分裂最小增益
        self.tree = None

    # 计算最优叶子权重 w*
    def get_leaf_weight(self, G, H):
        return -G / (H + self.reg_lambda)

    # 计算分裂增益 Gain
    def gain(self, G_L, H_L, G_R, H_R, G, H):
        gain = 0.5 * (G_L**2/(H_L+self.reg_lambda) + 
                      G_R**2/(H_R+self.reg_lambda) - 
                      G**2/(H+self.reg_lambda)) - self.gamma
        return gain

    # 分裂节点
    def split(self, X, g, h, depth):
        m, n = X.shape
        if depth >= self.max_depth:
            return self.get_leaf_weight(g.sum(), h.sum())

        best_gain = 0
        best_feat = None
        best_thresh = None
        G, H = g.sum(), h.sum()

        # 遍历所有特征找最优分裂
        for d in range(n):
            sorted_idx = np.argsort(X[:, d])
            for i in range(m - 1):
                idx = sorted_idx[:i+1]
                G_L, H_L = g[idx].sum(), h[idx].sum()
                G_R, H_R = G - G_L, H - H_L
                current_gain = self.gain(G_L, H_L, G_R, H_R, G, H)
                if current_gain > best_gain:
                    best_gain = current_gain
                    best_feat = d
                    best_thresh = (X[sorted_idx[i], d] + X[sorted_idx[i+1], d])/2

        if best_gain <= 0:
            return self.get_leaf_weight(G, H)

        # 分裂左右节点
        left = X[:, best_feat] <= best_thresh
        right = ~left
        return {
            "feat": best_feat,
            "thresh": best_thresh,
            "left": self.split(X[left], g[left], h[left], depth+1),
            "right": self.split(X[right], g[right], h[right], depth+1)
        }

    def fit(self, X, g, h):
        self.tree = self.split(X, g, h, depth=0)

    # 单样本预测
    def predict_one(self, x, tree):
        if not isinstance(tree, dict):
            return tree
        if x[tree["feat"]] <= tree["thresh"]:
            return self.predict_one(x, tree["left"])
        else:
            return self.predict_one(x, tree["right"])

    def predict(self, X):
        return np.array([self.predict_one(x, self.tree) for x in X])

# 完整 XGBoost 模型
class SimpleXGBoost:
    def __init__(self, n_estimators=10, learning_rate=0.1, max_depth=3, reg_lambda=1, gamma=0):
        self.n_estimators = n_estimators
        self.lr = learning_rate
        self.max_depth = max_depth
        self.reg_lambda = reg_lambda
        self.gamma = gamma
        self.trees = []
        self.init_pred = None

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        self.init_pred = np.mean(y)
        pred = np.full_like(y, self.init_pred, dtype=float)

        for _ in range(self.n_estimators):
            # 🔥 真正 XGBoost：计算一阶 & 二阶梯度 g, h
            g = pred - y
            h = np.ones_like(y)

            # 训练一棵 XGBTree
            tree = XGBTree(self.max_depth, self.reg_lambda, self.gamma)
            tree.fit(X, g, h)
            self.trees.append(tree)

            # 更新预测
            pred += self.lr * tree.predict(X)

    def predict(self, X):
        pred = np.full(len(X), self.init_pred)
        for tree in self.trees:
            pred += self.lr * tree.predict(X)
        return pred

# ------------------- 测试 -------------------
if __name__ == "__main__":
    from sklearn.datasets import load_diabetes
    from sklearn.metrics import mean_squared_error

    X, y = load_diabetes(return_X_y=True)
    X, y = X[:200], y[:200]

    model = SimpleXGBoost(n_estimators=40, learning_rate=0.1, max_depth=10)
    model.fit(X, y)
    print("MSE:", mean_squared_error(y, model.predict(X)))
```

## 八、LightGBM（Light Gradient Boosting Machine）

**LightGBM** 是微软推出的高效梯度提升树框架，是**XGBoost 的轻量化、高速化版本**。
在保持高精度的前提下，训练速度提升 10~20 倍，内存占用大幅降低，特别适合**大规模数据、高维稀疏数据**场景。

---

### 1. LightGBM 核心定位
LightGBM 属于**GBDT 系列**，与 XGBoost 目标函数一致，但**训练方式、树生长策略、特征处理方式**全面优化。
它是工业界、大数据场景下最常用的模型之一。

---

### 2. LightGBM vs XGBoost 核心差异
- **XGBoost**：预排序算法 + **Level-wise（逐层生长）**
- **LightGBM**：**Histogram 算法** + **Leaf-wise（按叶子生长）** + GOSS + EFB

---

### 3. 目标函数（与 XGBoost 完全一致）
#### 3.1 整体加法模型
$$
\hat{y}_i = \sum_{k=1}^K f_k(x_i)
$$

#### 3.2 第 t 轮目标函数
$$
Obj^{(t)} = \sum_{i=1}^n l(y_i, \hat{y}_i^{(t)}) + \sum_{k=1}^t \Omega(f_k)
$$

正则项：
$$
\Omega(f) = \gamma T + \frac{1}{2} \lambda \sum w_j^2
$$

- $T$：叶子节点数
- $w_j$：叶子权重
- $\gamma, \lambda$：正则系数

#### 3.3 二阶泰勒展开
$$
Obj^{(t)} \approx \sum_{i=1}^n \left[ g_i f_t(x_i) + \frac12 h_i f_t(x_i)^2 \right] + \Omega(f_t)
$$
- $g_i$：一阶梯度
- $h_i$：二阶梯度

---

### 4. LightGBM 四大核心创新（最重要）
#### 4.1 Histogram 算法（直方图算法）
**思想**：
将连续特征分桶离散化为**直方图**，用桶代替原始值，大幅减少计算与存储。

- 把特征值划分到若干 bins 中
- 只对 bin 边界计算分裂
- 速度提升、内存下降
- 分裂精度略有损失，但泛化性更好

**具体步骤**：
##### 第1步：分桶（把连续值变成离散桶）
例如一个特征：
[2.1, 3.4, 1.2, 5.6, 4.3, 2.9, 3.3, 4.9]

设定桶数 = 4
自动按范围分成 4 个区间：
- bin0：1.0 ~ 2.0
- bin1：2.0 ~ 3.0
- bin2：3.0 ~ 4.0
- bin3：4.0 ~ 6.0

每个样本不再存原始值，只存 **属于哪个 bin**。

---

##### 第2步：为每个 bin 统计梯度信息
每个 bin 存储两个值：
- **G = 该 bin 内所有样本的 g 之和**
- **H = 该 bin 内所有样本的 h 之和**

这就是直方图：
bin0: G=12.3, H=5.1；bin1: G=8.4, H=3.2；bin2: G=6.7, H=2.9；bin3: G=10.1, H=4.4

**只存这 2 个值，不再存所有样本！**

---

##### 第3步：分裂只需要遍历 bins
预排序要遍历 **几百几千个分裂点**
Histogram 只需要遍历 **几十个 bin**

例如：
在 bin1 后分裂 →
- 左 = bin0 + bin1
- 右 = bin2 + bin3

直接用 **G、H 累加** 计算增益！

---

##### 第4步：选择增益最大的 bin 作为分裂点
分裂点不再是精确值，而是 **bin 的边界**。

---

#### 4.2 Leaf-wise 树生长策略（最优叶子优先分裂）
**与 XGBoost Level-wise 最大区别**：
- Level-wise（XGBoost）：一层一层均匀生长
- Leaf-wise（LightGBM）：每次选择**增益最大的叶子**分裂

优点：
- 同样叶子数下，精度更高
- 收敛更快

需配合：
- `num_leaves` 限制复杂度
- `max_depth` 防止过深

---

#### 4.3 GOSS（Gradient-based One-Side Sampling）单边梯度采样
**核心思想**：
梯度大的样本对分裂影响更大 → **全部保留**
梯度小的样本影响小 → **随机采样，并放大权重**

步骤：
1. 取梯度最大的前 $a\%$ 样本
2. 小梯度样本随机采 $b\%$
3. 小梯度样本乘以放大系数 $\frac{1-a}{b}$

GOSS 大幅减少样本量，几乎不损失精度。

---

#### 4.4 EFB（Exclusive Feature Bundling）互斥特征捆绑
**思想**：
将**互斥特征**（几乎不同时非零，如 one-hot 特征）捆绑成一个特征（通过加偏移量），减少特征数量。

- 高维稀疏特征 → 低维稠密特征
- 不损失精度
- 训练速度大幅提升

---

### 5. 分裂增益与最优叶子权重
#### 5.1 最优叶子权重
$$
w_j^* = -\frac{\sum g_i}{\sum h_i + \lambda}
$$

#### 5.2 分裂增益（直方图增益）
$$
Gain = \frac12 \left[ \frac{G_L^2}{H_L+\lambda} + \frac{G_R^2}{H_R+\lambda} - \frac{G^2}{H+\lambda} \right] - \gamma
$$
与 XGBoost 一致，但基于直方图计算。

---

### 6. LightGBM 为什么更快？
1. **Histogram** 代替预排序，速度提升数倍
2. **GOSS** 减少样本量
3. **EFB** 减少特征量
4. **Leaf-wise** 分裂更高效
5. **直方图缓存命中更高**
6. 支持并行加速

---

### 7. LightGBM 核心优势总结
- 速度极快、内存极低
- 精度与 XGBoost 相当甚至更高
- 支持类别特征、缺失值、稀疏数据
- 支持 GOSS、EFB、并行训练
- 工业界大数据标准模型

```python
import numpy as np

class LGBTree:
    def __init__(self, num_leaves=31, max_bins=20, reg_lambda=1.0, gamma=0.0, goss_top=0.2, goss_sample=0.1):
        self.num_leaves = num_leaves
        self.max_bins = max_bins
        self.reg_lambda = reg_lambda
        self.gamma = gamma
        self.goss_top = goss_top
        self.goss_sample = goss_sample
        self.tree = None

    def get_weight(self, G, H):
        return -G / (H + self.reg_lambda)

    def gain(self, GL, HL, GR, HR):
        return 0.5 * (GL**2/(HL+self.reg_lambda) + GR**2/(HR+self.reg_lambda)) - self.gamma

    def build_histogram(self, x_col, g, h):
        bin_edges = np.linspace(x_col.min(), x_col.max(), self.max_bins + 1)
        bins = np.digitize(x_col, bin_edges) - 1
        hist = np.zeros((self.max_bins, 2))
        for b in range(self.max_bins):
            mask = bins == b
            hist[b, 0] = g[mask].sum()
            hist[b, 1] = h[mask].sum()
        return hist, bin_edges

    def goss_sampling(self, g, h):
        n = len(g)
        abs_g = np.abs(g)
        top_k = int(n * self.goss_top)
        top_idx = np.argsort(abs_g)[-top_k:]
        rest_idx = np.setdiff1d(np.arange(n), top_idx)
        sample_k = int(len(rest_idx) * self.goss_sample)
        sample_idx = np.random.choice(rest_idx, sample_k, replace=False)
        weight = np.ones(n)
        weight[sample_idx] = (1 - self.goss_top) / self.goss_sample
        return top_idx, sample_idx, weight

    def get_best_split(self, X, g, h):
        if len(X) <= 1:
            return None
        G, H = g.sum(), h.sum()
        best_gain = 0
        best_split = None

        for d in range(X.shape[1]):
            try:
                hist, edges = self.build_histogram(X[:, d], g, h)
            except:
                continue
            cum_G, cum_H = 0, 0
            for b in range(self.max_bins - 1):
                cum_G += hist[b, 0]
                cum_H += hist[b, 1]
                current_gain = self.gain(cum_G, cum_H, G - cum_G, H - cum_H)
                if current_gain > best_gain:
                    best_gain = current_gain
                    best_split = (d, edges[b+1], cum_G, cum_H, G-cum_G, H-cum_H)
        return best_split if best_gain > 0 else None

    def fit(self, X, g, h):
        top_idx, sample_idx, weight = self.goss_sampling(g, h)
        sample_mask = np.zeros(len(g), dtype=bool)
        sample_mask[top_idx] = True
        sample_mask[sample_idx] = True

        X_sampled = X[sample_mask]
        g_sampled = g[sample_mask] * weight[sample_mask]
        h_sampled = h[sample_mask] * weight[sample_mask]

        leaves = [{"X": X_sampled, "g": g_sampled, "h": h_sampled, "path": []}]

        while len(leaves) < self.num_leaves:
            best_gain = -1
            best_leaf = None
            best_split = None
            best_idx = -1

            for i, leaf in enumerate(leaves):
                split = self.get_best_split(leaf["X"], leaf["g"], leaf["h"])
                if split is None:
                    continue
                current_gain = self.gain(split[2], split[3], split[4], split[5])
                if current_gain > best_gain:
                    best_gain = current_gain
                    best_leaf = leaf
                    best_split = split
                    best_idx = i

            if best_idx == -1:
                break

            leaves.pop(best_idx)
            feat, thresh, GL, HL, GR, HR = best_split
            left_mask = best_leaf["X"][:, feat] <= thresh
            right_mask = ~left_mask

            if np.sum(left_mask) == 0 or np.sum(right_mask) == 0:
                leaves.append(best_leaf)
                continue

            leaves.append({
                "X": best_leaf["X"][left_mask],
                "g": best_leaf["g"][left_mask],
                "h": best_leaf["h"][left_mask],
                "path": best_leaf["path"] + [(feat, thresh, "left")]
            })
            leaves.append({
                "X": best_leaf["X"][right_mask],
                "g": best_leaf["g"][right_mask],
                "h": best_leaf["h"][right_mask],
                "path": best_leaf["path"] + [(feat, thresh, "right")]
            })

        tree = {}
        for leaf in leaves:
            node = tree
            for feat, thresh, direc in leaf["path"]:
                key = (feat, thresh)
                if key not in node:
                    node[key] = {"left": {}, "right": {}}
                node = node[key][direc]
            node["weight"] = self.get_weight(leaf["g"].sum(), leaf["h"].sum())
        self.tree = tree

    def predict_one(self, x, tree):
        while True:
            if "weight" in tree:
                return tree["weight"]
            key = next(iter(tree.keys()))
            feat, thresh = key
            tree = tree[key]["left"] if x[feat] <= thresh else tree[key]["right"]

    def predict(self, X):
        return np.array([self.predict_one(x, self.tree) for x in X])

class SimpleLightGBM:
    def __init__(self, n_estimators=10, learning_rate=0.1, num_leaves=31, max_bins=20):
        self.n_estimators = n_estimators
        self.lr = learning_rate
        self.num_leaves = num_leaves
        self.max_bins = max_bins
        self.trees = []
        self.init_pred = None

    def efb_bundle(self, X):
        n_samples, n_feats = X.shape
        if n_feats <= 1:
            return X
        offset = np.max(X) + 1
        bundled = np.zeros(n_samples)
        for i in range(n_feats):
            bundled += X[:, i] + i * offset
        return bundled.reshape(-1, 1)

    def fit(self, X, y):
        X = np.array(X, dtype=np.float32)
        y = np.array(y, dtype=np.float32)
        X = self.efb_bundle(X)
        self.init_pred = np.mean(y)
        pred = np.full_like(y, self.init_pred, dtype=np.float32)

        for _ in range(self.n_estimators):
            g = pred - y
            h = np.ones_like(g)
            tree = LGBTree(num_leaves=self.num_leaves, max_bins=self.max_bins)
            tree.fit(X, g, h)
            self.trees.append(tree)
            pred += self.lr * tree.predict(X)
        return self

    def predict(self, X):
        X = self.efb_bundle(np.array(X, dtype=np.float32))
        pred = np.full(len(X), self.init_pred)
        for tree in self.trees:
            pred += self.lr * tree.predict(X)
        return pred

if __name__ == "__main__":
    from sklearn.datasets import load_diabetes
    from sklearn.metrics import mean_squared_error
    X, y = load_diabetes(return_X_y=True)
    X, y = X[:200], y[:200]

    model = SimpleLightGBM(n_estimators=80, learning_rate=0.5, num_leaves=80, max_bins=16)
    model.fit(X, y)
    print("MSE:", mean_squared_error(y, model.predict(X)))
```

## 九、模型评估指标

### 一、二分类任务（必考核心）

前置混淆矩阵定义：TP（真正例）、TN（真负例）、FP（假正例）、FN（假负例）

1. 准确率（Accuracy）

含义与用处：衡量整体预测正确的样本占比，计算直观，仅适用于类别完全平衡场景，类别不平衡时结果极具误导性，不可单独作为核心指标。

$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

2. 精确率（Precision，查准率）

含义与用处：预测为正例的样本中，真实正例的占比，关注降低假阳性，适用于假阳性代价高的场景，如垃圾邮件检测、金融欺诈识别。

$$Precision = \frac{TP}{TP + FP}$$

3. 召回率（Recall，查全率/TPR）

含义与用处：真实正例样本中，被成功预测为正例的占比，关注降低假阴性，适用于漏检代价高的场景，如疾病筛查、危险物品检测。

$$Recall = \frac{TP}{TP + FN}$$

4. F1分数（F1 Score）

含义与用处：精确率和召回率的调和平均数，兼顾两者表现，类别不平衡场景首选综合指标，避免单看一项指标的偏差。

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
=\frac{2}{\frac{1}{Precision} + \frac{1}{Recall}}
$$

5. 马修斯相关系数（MCC）

含义与用处：适用于极度类别不平衡场景的最稳健指标，取值范围[-1,1]，1代表完美预测，-1代表完全反向预测，0代表随机预测，比F1更全面。

$$MCC = \frac{TP \times TN - FP \times FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}$$

6. AUC-ROC

含义与用处：ROC曲线下的面积，取值范围[0.5,1]，衡量模型对样本的整体排序能力，阈值无关，通用二分类评估指标，不受类别分布影响。

ROC曲线：以FPR为横轴，以TPR为纵轴，衡量二分类模型能力的曲线。

7. 对数损失（LogLoss，交叉熵损失）

含义与用处：评估模型预测概率的准确性，取值越小越好，常用于模型训练、早停，适用于输出概率的分类模型。

$$LogLoss = -\frac{1}{N} \sum_{i=1}^N \left[ y_i \log(p(y=y_i)) + (1-y_i) \log(1-p(y=y_i)) \right]$$

### 二、多分类任务

1. 宏F1（Macro-F1）

含义与用处：先计算每一类别的F1，再求算术平均，平等对待所有类别，适用于小类别样本同样重要的场景。

$$Macro\text{-}F1 = \frac{1}{K} \sum_{i=1}^K F1_i$$

2. 加权F1（Weighted-F1）

含义与用处：按各类别样本数量加权平均F1，考虑类别分布，贴合模型整体实际表现，是多分类最常用的F1指标。

$$Weighted\text{-}F1 = \sum_{i=1}^K \left( \frac{N_i}{N} \times F1_i \right)$$

### 三、回归任务

1. 平均绝对误差（MAE）

含义与用处：预测值与真实值绝对误差的平均值，对异常值鲁棒性强，结果直观，单位与原始数据一致。

$$MAE = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|$$

2. 均方误差（MSE）

含义与用处：预测值与真实值误差平方的平均值，对大误差惩罚较重，可导性好，是回归训练最常用的损失函数。

$$MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2$$

3. 均方根误差（RMSE）

含义与用处：MSE的平方根，回归任务最常用评估指标，单位与原始数据一致，便于直观解读误差大小。

$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2}$$

4. 决定系数（R²）

含义与用处：衡量模型对数据的拟合程度，取值范围(-∞,1]，越接近1拟合效果越好，小于0说明模型表现不如直接预测均值。

$$R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}$$

### 四、排序与推荐任务

1. 精确率@k（Precision@k）

含义与用处：Top-k推荐结果中，相关样本占比，衡量Top-k结果的精准度，适用于推荐、搜索场景。

$$Precision@k = \frac{\text{Top-k中相关样本数}}{k}$$

2. 召回率@k（Recall@k）

含义与用处：Top-k推荐结果中，相关样本占总相关样本的比例，衡量Top-k结果的覆盖度。

$$Recall@k = \frac{\text{Top-k中相关样本数}}{\text{总相关样本数}}$$

3. 归一化折损累积增益（NDCG@k）

含义与用处：排序任务核心指标，衡量Top-k结果的排序质量，考虑相关度权重和位置折损，取值越接近1越好。

$$NDCG@k = \frac{DCG@k}{IDCG@k}, \quad DCG@k = \sum_{i=1}^k \frac{2^{rel_i} - 1}{\log_2(i+1)}$$

### 五、目标检测任务

1. 交并比（IoU）

含义与用处：衡量预测框与真实框的重合度，判断检测是否有效，是目标检测所有指标的基础。

$$IoU = \frac{\text{预测框} \cap \text{真实框}}{\text{预测框} \cup \text{真实框}}$$

2. 平均精度（AP）

含义与用处：单类别PR曲线下的面积，衡量单类别检测精度，取值越接近1越好。

公式：单类别精确率-召回率曲线下围成的面积

3. 平均均值精度（mAP@0.5、mAP@0.5:0.95）

含义与用处：mAP@0.5为IoU=0.5时所有类别AP的平均值；mAP@0.5:0.95为COCO标准指标，IoU从0.5到0.95取多阈值AP均值，是目标检测最权威指标。

$$mAP = \frac{1}{K} \sum_{i=1}^K AP_i$$

### 六、图像分割任务

1. 平均交并比（mIoU）

含义与用处：语义分割核心指标，计算所有类别IoU的平均值，取值越接近1效果越好。

$$mIoU = \frac{1}{K} \sum_{i=1}^K \frac{TP_i}{TP_i + FP_i + FN_i}$$

2. Dice系数

含义与用处：医学图像分割常用指标，等价于分割场景下的F1分数，侧重区域重合度，对小目标更友好。

$$Dice = \frac{2 \times |A \cap B|}{|A| + |B|}$$


## 十、过拟合解决方法

### 1. 数据层面（最根本，优先级最高）

- **收集更多真实数据**  
  最直接、最有效的办法。数据量增加，模型难以记住噪声。

- **数据增强（Data Augmentation）**  
  图像：翻转、旋转、裁剪、颜色抖动、Mixup、CutMix、RandAugment  
  文本：同义词替换、随机插入/删除、回译  
  时序：加噪声、时间窗滑动、窗口翻转  

- **类别不平衡处理**  
  重采样（过采样少数类 / 欠采样多数类）、SMOTE、ADASYN  
  损失函数加权（class_weight）、Focal Loss

### 2. 模型复杂度控制

- **减小模型容量**  
  减少层数、神经元数、树深度、叶子节点数  
  面试话术：“我先尝试把模型变小，看验证集是否提升，再逐步加复杂性。”

- **Early Stopping**  
  监控验证集指标，连续 N 次（patience）不下降就停止  
  几乎所有深度学习任务必提

- **正则化**  
  L2 正则（权重衰减 weight_decay）——最常用  
  L1 正则（产生稀疏性，用于特征选择）  
  Dropout（深度网络，p=0.1~0.5）  ：随机丢掉一些神经元
  DropPath / Stochastic Depth（ResNet / Transformer 深层网络）

- **Batch Normalization / Layer Normalization**  
  BN 本身有轻微正则效果，现代网络几乎必加

### 3. 集成学习

- **Bagging**（降低方差）  
  代表：Random Forest  
  多个独立模型投票/平均

- **Boosting**（降低偏差+方差）  
  代表：XGBoost / LightGBM / CatBoost  
  逐步修正残差 + 强正则化（学习率、树深度、正则项、早停）

- **模型融合**  
  Stacking / Blending / 加权平均  
  面试话术：“我通常会训练多个不同结构/初始化的模型，然后做加权融合或 Stacking。”

## 4. 损失函数与训练技巧（现代深度学习常考）

- **Label Smoothing**  
  把硬标签软化（如 [1, 0 ,0] --> [0.9, 0.05, 0.05]），现代大模型几乎必用

- **Mixup / CutMix / Mosaic**  
  样本/标签线性混合或区域混合，图像任务效果显著

- **学习率策略**  
  Warm-up + Cosine Decay  
  ReduceLROnPlateau / CosineAnnealing

- **Test-Time Augmentation (TTA)**  
  测试时多增强取平均

## 十二、bagging（Bootstrap Aggregating）

**Bagging** 是最早提出的集成学习方法之一（1996 年 Leo Breiman 提出），全称 **Bootstrap Aggregating**。  
它是“**并行式集成学习**”的代表，与 Boosting（串行）形成鲜明对比。

### 1. Bagging 核心思想

**“用有放回随机抽样生成多个不同的训练子集，训练多个独立的同类基模型，最后通过投票（分类）或平均（回归）来降低方差”**。

- **主要目标**：**降低模型的方差**（variance），对高方差模型（如决策树）效果特别显著
- **不降低偏差**（bias），甚至可能略微增加一点偏差
- **并行训练**：所有基模型可以完全并行，速度快

### 2. Bagging 的数学原理与关键步骤

#### 2.1 Bootstrap 采样（有放回抽样）

从原始数据集 D（大小 N）中，**有放回随机抽取 N 个样本**，生成一个子数据集 Dᵢ。

- 数学期望：每个样本被抽中的概率 ≈ 1 - (1 - 1/N)^N ≈ 1 - 1/e ≈ 63.2%
- 约 36.8% 的原始样本不会出现在某个子数据集中 → 这些样本称为 **Out-of-Bag (OOB) 样本**（天然的验证集）

#### 2.2 训练多个独立基模型

对 B 个 Bootstrap 采样得到的子数据集 $D₁, D₂, ..., D_B$，分别训练 B 个基模型 $h₁, h₂, ..., h_B$。

基模型通常选择**高方差、低偏差**的模型，最经典是**未剪枝的决策树**。

#### 2.3 预测方式

**分类任务**（多数投票）：

$$
\hat{y}(x) = \text{mode} \Big\{ h_1(x), h_2(x), \dots, h_B(x) \Big\}
$$

**回归任务**（简单平均）：

$$
\hat{y}(x) = \frac{1}{B} \sum_{b=1}^B h_b(x)
$$

#### 2.4 为什么能降低方差？（数学解释）

假设基模型的期望偏差为 bias，单个模型方差为 σ²。

Bagging 后：

- 期望偏差 ≈ bias（几乎不变）
- 方差 ≈ σ² / B（假设基模型之间完全独立）

→ 方差下降约 B 倍（实际因为 Bootstrap 相关性，下降不到 B 倍，但仍然显著）

### 3. Bagging 与随机森林的关系

**随机森林 = Bagging + 特征随机子采样**

| 项目                  | 纯 Bagging（决策树）          | 随机森林（Random Forest）             |
|-----------------------|--------------------------------|----------------------------------------|
| 采样方式              | Bootstrap                      | Bootstrap                              |
| 每棵树特征选择        | 全部特征                       | 每个节点随机选 m 个特征（m << M）      |
| 主要降方差来源        | 样本随机性                     | 样本随机性 + 特征随机性                |
| 性能                  | 很好                           | 通常明显更好                           |
| 代表实现              | sklearn BaggingClassifier      | sklearn RandomForestClassifier         |

## 十三、boosting

**Boosting** 是机器学习中最强大的集成学习范式之一，与 Bagging（并行降低方差）不同，Boosting 是**串行**的、**逐步修正残差**的**自适应加权**方法，主要目标是**同时降低偏差和方差**，通常能获得极高的预测精度。

### 1. Boosting 核心思想（一句话记住）

**“一棵树接着一棵树地学习，每一棵新树都重点关注前一棵树做错的样本（残差或加权误差），最终将所有弱学习器加权累加成强学习器”**。

- 核心机制：**逐步聚焦难样本** + **加权投票/加权平均**
- 代表算法演进顺序（重要性由高到低）：
  1. AdaBoost（自适应提升） → 最早经典
  2. Gradient Boosting（梯度提升） → 现代基础
  3. XGBoost / LightGBM / CatBoost → 工业界杀手级实现

### 2. Boosting 的三大经典阶段对比

### 3. Boosting 通用数学框架（以梯度提升为例）

#### 3.1 加法模型

最终模型：

$$
F_M(x) = F_0(x) + \sum_{m=1}^M \eta \cdot f_m(x)
$$

- $ F_0(x) $：初始预测（回归通常是 y 的均值，分类是 log-odds）
- $ \eta $：学习率（shrinkage，0.01~0.3）
- $ f_m(x) $：第 m 棵树（弱学习器）

#### 3.2 第 m 步的目标（梯度提升）

最小化伪残差：

$$
r_{i,m} = -\left[ \frac{\partial L(y_i, F(x_i))}{\partial F(x_i)} \right]_{F=F_{m-1}}
$$

然后拟合一棵树 $ f_m(x) $ 去逼近残差 $ r_{i,m} $

#### 3.3 更精确的现代形式（XGBoost / LightGBM）

第 m 步目标函数（带正则）：

$$
Obj^{(m)} = \sum_{i=1}^n l(y_i, F_{m-1}(x_i) + f_m(x_i)) + \Omega(f_m)
$$

用二阶泰勒展开近似：

$$
Obj^{(m)} \approx \sum_{i=1}^n \left[ g_i f_m(x_i) + \frac{1}{2} h_i f_m(x_i)^2 \right] + \Omega(f_m)
$$

其中：

- $ g_i = \partial l / \partial F_{m-1} $ （一阶梯度）
- $ h_i = \partial^2 l / \partial F_{m-1}^2 $ （二阶梯度）
- $ \Omega(f_m) = \gamma T + \frac{1}{2} \lambda \sum w_j^2 $ （叶子数惩罚 + L2 正则）
