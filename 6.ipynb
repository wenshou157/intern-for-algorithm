```python
import torch 
import matplotlib.pyplot as plt 
```

```python
x = torch.arange(-7, 7, 0.01)
def normal(x, mu, sigma):
    p = 1 / torch.sqrt(2 * torch.tensor(torch.pi) * sigma**2)
    return p * torch.exp(-0.5 * (x - mu)**2 / sigma**2 )  
params = [(0, 1), (0, 2), (3, 1)]
plt.figure(figsize = (4.5, 2.5))
styles = ['b-', 'm--', 'g:']
for i, (mu, sigma) in enumerate(params):
    plt.plot(x.numpy(), normal(x, mu, sigma).numpy(), styles[i], label = f'mean {mu}, std {sigma}')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```
